```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

Data = {'Name': ["Jefferson", "Derry", "Dorry", "Clency"],
        'Class': [1,2,3,4],
        'Garlic_Consumption': [7, 5, 2,3],
}

labels = ['Jefferson', 'Derry', 'Dorry', 'Clency']
plt.pie(Data['Garlic_Consumption'], labels=labels)
plt.title('Garlic Consumption')
plt.show()

df = pd.DataFrame(Data)
print(df)

plt.plot(df['Garlic_Consumption'])
plt.title('Garlic Consumption')
plt.xlabel('Name')
plt.ylabel('Garlic Consumption')
plt.grid(True)
plt.show()
```
